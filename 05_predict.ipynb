# Predict

Loads models, vocabulary, and makes prediction for all files in `predict_images_path`

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

output_folder_path = f'{drive_path}model_output/'
predict_images_path = f'{drive_path}data/predict_images/'

# checkpoint = torch.load(f'{output_folder_path}BEST_checkpoint_flickr8k_5_cap_per_img_5_min_word_freq_v001.pth.tar')
checkpoint = torch.load(f'{output_folder_path}BEST_checkpoint_flickr8k_5_cap_per_img_5_min_word_freq.pth.tar')
with open(f'{output_folder_path}WORDMAP_flickr8k_5_cap_per_img_5_min_word_freq.json', 'r') as file:
  wordmap = json.load(file)

decoder = checkpoint['decoder']
encoder = checkpoint['encoder']
vocabulary = list(wordmap.keys())

directory = os.fsencode(predict_images_path)

    
for file in os.listdir(directory):

  filename = os.fsdecode(file)

  caption = caption_image_beam_search(
      encoder,
      decoder,
      f'{predict_images_path}{filename}',
      wordmap,
      beam_size=3
      )
  
  caption_list = []

  for key in caption[0]:
    caption_list.append(vocabulary[key-1])

  # image = Image.open(f'{predict_images_path}51155257684_f2bbcd7571_w.jpg')
  image = Image.open(f'{predict_images_path}{filename}')
  plt.figure(figsize=(8, 8))
  plt.title(' '.join(caption_list[1:-1]))
  plt.imshow(image)
  #print(' '.join(caption_list[1:-1]))
```
